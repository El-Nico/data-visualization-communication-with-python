```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from __future__ import print_function 
%matplotlib inline 
```

```python
#pip install seaborn
```

```python
import seaborn as sns
```

```python
datasetUrl = 'https://raw.githubusercontent.com/El-Nico/data-visualization-communication-with-python/main/Dataset_Seaborn/Body_fat.csv'
dataset = pd.read_csv(datasetUrl) 
dataset.head()
```

```python
#Plot to see the distribution of a variable
sns.displot(dataset['Abdomen'], color = 'green')
```

```python
#Scatter plot
sns.jointplot(x='Abdomen', y='BodyFat', data=dataset, kind='reg', color='orange')

#Changing the size
sns.set_style('darkgrid')
plt.figure(figsize=(8,4))
sns.jointplot(x='Abdomen', y='BodyFat', data=dataset, kind='reg', color='orange')
```

```python
#Distribution plot: Area plot
sns.jointplot(x='Abdomen', y='BodyFat', data=dataset, kind='kde')
plt.title('Abdomen fat vs Body fat', color='blue', fontsize=15)
```

```python
#Distribution plot: Hexagonal plot
sns.jointplot(x='Abdomen', y='BodyFat', data=dataset, kind='hex', color='purple')
```

```python
#KDE PLOT
sns.kdeplot(dataset['Chest'])
plt.title('KDE Plot', fontsize=15)
```

```python
#PAIR PLOT
sns.pairplot(dataset)
```

```python
#BAR PLOT
sns.barplot(x='Gender', y='BodyFat', data=dataset, estimator=np.std)#/median/std/var)
plt.ylabel('Body fat StandardDeviation')
plt.title('Male vs Female')
```

```python
#BAR PLOT
sns.barplot(x='Gender', y='BodyFat', data=dataset) #estimator=np.mean/median/std/var)
plt.title('Male vs Female')
```

```python
#BOX PLOT
sns.boxplot(x='Day', y='BodyFat', data=dataset, hue='Gender')
plt.title('Box plot', fontsize=15)
```

```python
#VIOLIN PLOT
sns.violinplot(x='Day', y='BodyFat', data=dataset, hue='Gender', split=True)
plt.title('Violin plot')
```

```python
#Strip plot
plt.figure(figsize=(8,5))
sns.stripplot(x='Day', y='BodyFat', data=dataset, jitter=True, hue='Gender') #dodge=True)
plt.title('Strip Plot', fontsize=15)
```

```python
#HEATMAP
#create the correlation table
plt.figure(figsize=(8,6))
sns.set_context('paper', font_scale=1.4)
mx= dataset.corr()
mx
```

```python
sns.heatmap(mx, annot=True, cmap='Oranges')
```
