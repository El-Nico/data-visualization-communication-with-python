```python
#GETTING STARTED WITH PLOTS

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# define two variables,  x and y
x = [1, 2, 3, 4, 5]
y = [1, 4, 6, 8, 10]
plt.plot(x, y)
```

```python
#Add the labels to the original graph
# Plot on that set of axes
plt.plot(x, y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('Title Here')
```

```python
#change the color of the line
plt.plot(x, y, 'r')
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('Title Here')
plt.show()
```

```python
#change the form of the line
plt.plot(x, y, 'g--')
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('Title Here')
plt.show()
```

```python
# change the form of the line
plt.plot(x, y, 'b*-')
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('Title Here')
plt.show()
```

```python
#create multiple plots on the same canvas

# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

```python
#Let's try with some real example

# Importing the dataset
#dataset = pd.read_csv('LOCATION_OF_YOUR_FILE)/NAME_OF_YOUR_FILE.CSV')

dataset = pd.read_csv('https://raw.githubusercontent.com/El-Nico/data-visualization-communication-with-python/main/Datasets-20210929/Salaries.csv')
```

```python
#Visualise the 5 first rows of the file

dataset.head(5)
```

```python
adm = dataset[dataset.Position == 'Administration']
```

```python
adm
```

```python
#plot it
plt.plot(adm.Year, adm.Salary)
```

```python
#Add the labels and features
plt.plot(adm.Year, adm.Salary, 'g--')
plt.xlabel('Year')
plt.ylabel('Salary')
plt.title('Administration Salary over the years')
```

```python
#Business Intelligence

bi = dataset[dataset.Position == 'Business Intelligence']
```

```python
bi
```

```python
#plot it!
plt.plot(bi.Year, bi.Salary, 'b--')
plt.xlabel('Year')
plt.ylabel('Salary')
plt.title('BI Salary over the years')
```

```python
#compare the Administration and BI salaries

plt.plot(adm.Year, adm.Salary)
plt.plot(bi.Year, bi.Salary)
plt.xlabel('Year')
plt.ylabel('Salary')
plt.title('Administration vs BI')
plt.legend(['Administration', 'BI'])
```

```python
#Marketing

mkt = dataset[dataset.Position == 'Marketing']
```

```python
mkt
```

```python
#compare the BI and MKT salaries

plt.figure(figsize=(15,10))
plt.plot(bi.Year, bi.Salary)
plt.plot(mkt.Year, mkt.Salary)
plt.xlabel('Year')
plt.ylabel('Salary')
plt.title('BI vs MKT')
plt.legend(['BI', 'MKT'])
```

```python
#BAR CHART
plt.figure(figsize=(10,8))
plt.xlabel('Year')
plt.ylabel('Salary')
plt.title('MKT Salary over the years', fontsize = 15)
plt.bar(x,y, color= 'purple')
```

# 50 Startups Dataset

```python
dataset= pd.read_csv('https://raw.githubusercontent.com/El-Nico/data-visualization-communication-with-python/main/Datasets-20210929/50_Startups.csv')
dataset.head(10)
```

###Line plot comparing R&D and Administration

```python
adm = dataset['Administration']
rd= dataset['R&D Spend']
profit= dataset['Profit']
rd, adm
```

```python
#compare the R&D and Administration against profit

plt.plot(rd, profit)
plt.plot(adm, profit)
plt.xlabel('Profit')
plt.ylabel('Feature')
plt.title('Effect of R&D vs Administration')
plt.legend(['R&D', 'Administration'])
```
