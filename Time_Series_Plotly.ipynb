```python
#Importing libraries

import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading data from GitHub

dataset = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/finance-charts-apple.csv')
dataset
```

```python
fig = go.Figure([go.Scatter(x=dataset['Date'], y=dataset['AAPL.Open'])])
fig.show()
```

```python
#Importing another dataset from GitHub

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master//tesla-stock-price.csv')
#Open
#High
#Low
df
```

```python
df.sort_values('date', inplace=True)
```

```python
fig = go.Figure([go.Scatter(name = 'Open', x=df['date'], y=df['open'], mode='markers', text=df['close'], )]) 
fig.add_scatter(name = 'High',x=df['date'], y=df['high'], mode='lines')
fig.add_scatter(name = 'Low',x=df['date'], y=df['low'], mode='lines')
fig.update_layout(title='Tesla Stock Performance 2 year outdated')
fig.show()
```

```python
#Bar charts including timeline with plotly

dataframe = px.data.gapminder()
dataframe
```

```python
df_italy = px.data.gapminder().query("country == 'Italy'")
fig = px.bar(df_italy, x='year', y='pop')
fig.update_layout(title_text="Population in Italy")
fig.show()
```

```python
#More colors!

fig = px.bar(df_italy, x='year', y='pop', hover_data=['continent','gdpPercap'], color='lifeExp',labels={'Population/Life Exp of Italy'}, height=600)
fig.update_layout(title_text="Population in Italy")
fig.show() 
```
